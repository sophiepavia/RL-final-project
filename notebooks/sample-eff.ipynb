# Sample Efficiency

```python
path1 = '../bin/logs/a2c_Pendulum-v1/sample_efficiency.csv'
path2 = '../bin/logs/td3_Pendulum-v1/sample_efficiency.csv'
path3 = '../bin/logs/sac_Pendulum-v1/sample_efficiency.csv'

# Determine the amount of experience required to achieve a certain level of performance
performance_threshold = -250 # Pendulum-v1 environment has a maximum reward of 0, this value indicates pendulum is being controlled reasonbly well though not perfectly upright
```

```python
path1 = '../bin/logs/a2c_BipedalWalker-v3/sample_efficiency.csv'
path2 = '../bin/logs/td3_BipedalWalker-v3/sample_efficiency.csv'
path3 = '../bin/logs/sac_BipedalWalker-v3/sample_efficiency.csv'

# Determine the amount of experience required to achieve a certain level of performance
performance_threshold = 250 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV files
data1 = pd.read_csv(path1)
data2 = pd.read_csv(path2)
data3 = pd.read_csv(path3)

# Plot the data
plt.figure(figsize=(12, 8))
plt.plot(data1['timesteps'], data1['rewards'], label='Algorithm 1')
plt.plot(data2['timesteps'], data2['rewards'], label='Algorithm 2')
plt.plot(data3['timesteps'], data3['rewards'], label='Algorithm 3')
plt.xlabel('Timesteps')
plt.ylabel('Rewards')
plt.title('Sample Efficiency Comparison of RL Algorithms')
plt.legend()
plt.grid(True)
plt.show()

# Calculate sample efficiency metrics for all three algorithms
average_reward1 = data1['rewards'].mean()
max_reward1 = data1['rewards'].max()
min_reward1 = data1['rewards'].min()

average_reward2 = data2['rewards'].mean()
max_reward2 = data2['rewards'].max()
min_reward2 = data2['rewards'].min()

average_reward3 = data3['rewards'].mean()
max_reward3 = data3['rewards'].max()
min_reward3 = data3['rewards'].min()

print(f"Algorithm 1 - Average Reward: {average_reward1}, Max Reward: {max_reward1}, Min Reward: {min_reward1}")
print(f"Algorithm 2 - Average Reward: {average_reward2}, Max Reward: {max_reward2}, Min Reward: {min_reward2}")
print(f"Algorithm 3 - Average Reward: {average_reward3}, Max Reward: {max_reward3}, Min Reward: {min_reward3}")

timesteps_to_threshold_1 = data1[data1['rewards'] >= performance_threshold]['timesteps'].min()
timesteps_to_threshold_2 = data2[data2['rewards'] >= performance_threshold]['timesteps'].min()
timesteps_to_threshold_3 = data3[data3['rewards'] >= performance_threshold]['timesteps'].min()

print(f"Algorithm 1 - Timesteps to reach performance threshold ({performance_threshold}): {timesteps_to_threshold_1}")
print(f"Algorithm 2 - Timesteps to reach performance threshold ({performance_threshold}): {timesteps_to_threshold_2}")
print(f"Algorithm 3 - Timesteps to reach performance threshold ({performance_threshold}): {timesteps_to_threshold_3}")
```
